##### visualizing the distribution of a dataset
##### %matplotlib inline indicate print everytime plotting is done

```python
import pandas 
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

##### importing automobile dataset

```python
automobile_dataset = pandas.read_csv("Automobile (1).csv")
```

```python
automobile_dataset
```

```python
automobile_dataset.head(20)
```

##### plotting univariate distributions

```python
sns.distplot(automobile_dataset['normalized_losses'])
```

```python
sns.distplot(automobile_dataset['engine_size'])
```

```python
sns.distplot(automobile_dataset['city_mpg'])
```

##### histograms

```python
sns.distplot(automobile_dataset['city_mpg'], kde=False, rug=True);
```

##### univariate distribution

```python
sns.distplot(automobile_dataset['normalized_losses'], kde=False, rug=True);
```

##### bivariate distribution

```python
sns.jointplot(x=automobile_dataset['engine_size'], y=automobile_dataset['horsepower'])	
```

##### hexbin plots

```python
#kind: {"scatter" | "reg" | "resid" | "kde" | "hex"}
sns.jointplot(x=automobile_dataset['engine_size'], y=automobile_dataset['horsepower'], kind="hex")	
```

```python
#kind: {"scatter" | "reg" | "resid" | "kde" | "hex"}
sns.jointplot(x=automobile_dataset['engine_size'], y=automobile_dataset['horsepower'], kind="reg")	
```

```python
#kind: {"scatter" | "reg" | "resid" | "kde" | "hex"}
sns.jointplot(x=automobile_dataset['engine_size'], y=automobile_dataset['horsepower'], kind="kde")	
```

##### multivariate distribution
##### visualizing pairwise relationships in a dataset

```python
sns.pairplot(automobile_dataset[['normalized_losses', 'engine_size', 'horsepower']])
```

##### plotting with categorical data
##### striplots

```python
sns.stripplot(x=automobile_dataset['fuel_type'], y=automobile_dataset['horsepower'], jitter= True)
```

##### swarmplot

```python
sns.swarmplot(x=automobile_dataset['fuel_type'], y=automobile_dataset['horsepower'])
```

##### boxplots

```python
sns.boxplot(x=automobile_dataset['number_of_doors'], y=automobile_dataset['horsepower'])
```

```python
sns.countplot(x=automobile_dataset['body_style'], hue=automobile_dataset['engine_location'])
```

##### boxplots

```python
sns.boxplot(x=automobile_dataset['number_of_doors'], y=automobile_dataset['horsepower'], hue=automobile_dataset['fuel_type'])
```

##### barplots
##### lines on top of the barplot is confidence intervals

```python
sns.barplot(x=automobile_dataset['body_style'], y=automobile_dataset['horsepower'], hue=automobile_dataset['engine_location'])
```

##### pointplots

```python
sns.pointplot(x=automobile_dataset['fuel_system'], y=automobile_dataset['horsepower'], hue=automobile_dataset['number_of_doors'])
```

##### catplot does same as factorplot which is complex type of plotting
##### drawing muti-panel categorical plots

```python
sns.catplot(x="fuel_type",
              y="horsepower",
              hue="number_of_doors",
              col="engine_location",
              data=automobile_dataset,
              kind="swarm")

#various types of kind input: {point, bar, count, box, violin, strip)
```

```python
sns.catplot(x="fuel_type",
              y="horsepower",
              hue="number_of_doors",
              col="engine_location",
              data=automobile_dataset,
              kind="violin")

#various types of kind input: {point, bar, count, box, violin, strip)
```

##### function to draw linear regression models

```python
sns.lmplot(x="horsepower",y="peak_rpm", data=automobile_dataset)
```

```python
sns.lmplot(x="horsepower",y="peak_rpm", data=automobile_dataset, hue="fuel_type")
```
